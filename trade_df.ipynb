## 주말을 분류하여 trade_day_devide에 1로 표기한다.(나머지는 0)

### 먼저 0만들어 있는 새로운 열을 하나 만든 뒤 trade_day에서 3,4,5인 날만 확인 후 해당 인자를 변경한다.

```python
trade_df['trade_day_devide']=np.zeros(trade_df.shape[0])
```

```python
for i in range(trade_df.shape[0]):
    if trade_df.trade_day[i] in (3,4,5):
        trade_df['trade_day_devide']=1
    else:
        trade_df['trade_day_devide']=0
```

## item type의 카테고리를 열로 추가한다.

### get_dummies를 통해 아이템의 카테고리들을 열로 바꾸고 index를 리셋시킨다.

```python
dummy=pd.get_dummies(trade_df.item_type).reset_index()
```

```python
dummy=dummy.drop(['index'],axis=1)
```

```python
dummy.shape
```

## 더미함수와 원 데이터를 결합한다.

```python
pd.concat([trade_df,dummy],axis=1)
```
